```python
numeros=(15, 8, 42, 3, 27, 19, 5)
min_num=min(numeros)
max_num=max(numeros)

diff=max_num-min_num
print(f'el maximo numero es: {max_num}\nel minimo numero es: {min_num}\nla diferencia es: {diff}')
```

```python
#Ejercicio 3
nombres = ('Ana','Luis','Pedro','María','Luis','Carlos')
nom=input('Escribe un nombre: ')
if nom in nombres:
    print(f'La numeracion que corresponde al nombre {nom} es: ', nombres.index(nom)+1)

else:
    print('El nombre no existe en la lista')
```

```python
#Ejercicio 4 
notas = (18,15,20,11,14,17,19)
a=max(notas)
b=min(notas)
c=round(sum(notas)/len(notas),3)
for nota in notas:
    if nota<=11:
        print('La nota desaprobada es:',nota)
print(f'La maxima nota es: {a}, la minima nota es: {b}, y el promedio es: {c}')
```
