```python
mi_diccionario={
    'peru':'sol',
    'japon':'yen',
    'usa':'dolar americano'
}
print(mi_diccionario)
```

```python
mi_diccionario['mexico']='peso mexicano'
print(mi_diccionario)
```

```python
# update permite actualizar incluso una nueva llave
mi_diccionario['peru']='nuevo sol'
print(mi_diccionario)
mi_diccionario.update({'japon':'yen japones'})
print(mi_diccionario)
mi_diccionario.update({'china':'yuan'})
```

```python
mi_diccionario.keys()
```

```python
mi_diccionario.values()
```

```python
print(mi_diccionario.get('peru'))
# get da la opcion de devolver un valor en caso no este encontrado
print(mi_diccionario.get('argentina','Buenos Aires'))
```

```python
#los corchetes si te pueden retornar error en caso no haya el elemento
mi_diccionario['peru']
```

```python
mi_diccionario.setdefault('bolivia','pesos bolivianos')
print(mi_diccionario)
```

```python
mi_diccionario.pop()
print(mi_diccionario)
```

```python
mi_diccionario={
    'peru':'lima',
    'moruega':'oslo'
}
print(len(mi_diccionario))
```

```python
for key in mi_diccionario:
    print(key)
    print(mi_diccionario[key])
    print('='*10)
    
```

```python
for key,value in mi_diccionario.items():
    print(key)
    print(value)
    print('='*10)
```
