```python
#Ejercicio 1
ventas= [150.50, 200, 80.90, 300, 175.50, 90] 

def analizar_ventas(ventas:list[float|int])->dict[str,int|float]: #->dict[str|float] indica que retornara un valor flotante
    
    if not ventas:
        return 'No hay ventas'
    for venta in ventas:
        if venta<0:
            return f'Hay ventas con valores negativas: {venta}'

    total_vendido=sum(ventas)
    promedio_ventas=sum(ventas)/len(ventas)
    maxima_venta=max(ventas)
    minima_venta=min(ventas)
    ventas_mayores=0
    ventas_menores=0
    
    for valor in ventas:
        if valor>promedio_ventas:
            ventas_mayores=ventas_mayores+1

        if valor<100:
            ventas_menores=ventas_menores+1
    
    return {
        'total_vendido':total_vendido,
        'promedio_ventas':promedio_ventas,
        'venta maxima':maxima_venta,
        'venta minima':minima_venta,
        'ventas mayores a promedio':ventas_mayores,
        'ventas menores a 100':ventas_menores

    }



analizar_ventas(ventas)
```

```python

simbolos_need=('@',',', '#', '$', '%')

def validar_password(password:str|float|int)->tuple[bool,list[str]]:
    errores=[]
    simbolos_need=('@',',', '#', '$', '%')

    if len(password)<8:
            errores.append('La contraseña tiene muy pocos caracteres.')

    contiene_minuscula=False
    contiene_mayuscula=False
    contiene_numero=False
    contiene_simbolos=False

    for caracter in password:
        if caracter.islower():
            contiene_minuscula=True
            break
    
    for caracter in password:
        if caracter.isupper():
            contiene_mayuscula=True
            break
    
    for caracter in password:
        if caracter.isnumeric():
            contiene_numero=True
            break

    for caracter in password:
        if caracter in simbolos_need:
            contiene_simbolos=True
            break


    if not contiene_minuscula:
         errores.append('La contraseña no contiene minuscula')

    if not contiene_mayuscula:
         errores.append('La contraseña no contiene mayuscula')

    if not contiene_numero:
         errores.append('La contraseña no contiene numero')

    if not simbolos_need:
         errores.append('La contraseña no contiene ninguno de los simbolos requeridos')
    
    if len(password)!=len(password.strip()):
         errores.append('La contraseña contiene espacios en blanco')

    es_valida=False if errores else True
    return(es_valida,errores)


es_valida, errores = validar_password('Python123') 
print(es_valida) 
print(errores)
```

```python
#Ejercicio 4
palabras = [ 
'python', 'Java', 'código', 'web', 
'java', 'datos', 'API', 'lista' 
] 

def agrupar_por_longitud(palabras:list[str])->dict[int,list[str]]:
    palabras_limpias=set([
            palabra.lower()
            for palabra in palabras
            if isinstance(palabra,str)
    ])
    grupos={}
    for palabras_limpias in palabras_limpias:
        longitud=len(palabras_limpias)
        grupos.setdefault(longitud,[]).append(palabras_limpias) #setdefault crea una nueva lista en caso no exista
    
    print(grupos)
    
    for llave in grupos:
        grupos[llave].sort()
    
    print(grupos)

    return(grupos)

print(agrupar_por_longitud(palabras))
agrupar_por_longitud(palabras)
```

```python
inventario = { 
'Laptop': 5, 
'Mouse': 12, 
'Teclado': 8 
} 

def actualizar_inventario(
        inventario:dict[str,int], 
        producto:str, 
        cantidad:int, 
        operacion:str
        )->dict[str,str|int]:
    
    operaciones=('entrada','salida')
    operaciones=operacion.lower()

    if operacion not in operaciones:
        return {
            'mensaje':f'Error, tipo  de operacion  no soportada:{operacion}',
            'stock actual': f'{inventario[producto]}'
        }
    if not isinstance(cantidad,int):
        return {
            'mensaje':f'Error: tipo de dato inadecuado se esperaba int y se obtuvo {type(cantidad)}'
        }
    
    if cantidad<=0:
        return {
            'mensaje':'Error: el dato es negativo'
        }
    
    if operacion=='entrada':
        #inventario.setdefault(producto,cantidad)
        if not producto in inventario:
            inventario[producto]=cantidad
        else:
            inventario[producto]+=cantidad

    if operacion=='salida' and producto in inventario and inventario[producto]>=cantidad:
        inventario[producto]=inventario[producto]-cantidad
    return {
        'Comprobacion exitosa'
        'stock actual':f'{inventario[producto]}'
    }

actualizar_inventario(inventario, 'Mouse', 520, 'entrada')
print(inventario)
```

```python
#Ejercicio 8
texto = 'aaabbccccdaaffff' 
#comprimido = comprimir_texto(texto) 
#original = descomprimir_texto(comprimido)

def comprimir_texto(texto:str)->str:
    resultado=[]
    caracter=texto[0] #El [0] indica la primera posicion de texto
    contador=1

    for letra in texto[1:]: #Se coloca [1] debido a que el [0] ya esta como inicio, este continua
        if letra==caracter:
            contador+=1
        else:
            resultado.append(f'{caracter}{contador}')
            caracter=letra
            contador=1
    resultado.append(f'{caracter}{contador}')
    
    resultado=''.join(resultado)

    return resultado

comprimir_texto(texto)    

def descomprimir_texto(texto:str)->str:
    letras=texto[0::2]
    numeros=texto[1::2]
    resultado=[]

    for letra,numero in zip(letras,numeros):
        resultado.append(
            letra*int(numero)
        )
    resultado=''.join(resultado) #Sirve para concatenar elementos
    return(resultado)
comprimir_texto(texto)
```
