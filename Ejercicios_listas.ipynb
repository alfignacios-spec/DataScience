```python
#Ejercicio 8
lista1 = [1, 3, 5, 7, 9, 11] 
lista2 = [2, 3, 6, 7, 10, 11]

no_comunes=[]
for num in lista1:
    if num not in lista2:
        no_comunes.append(num)

for num in lista2:
    if num not in lista1:
        no_comunes.append(num)

comunes=[]
for num in lista1:
    if num in lista2:
        comunes.append(num)

print(f'los elementos semejantes son:{comunes}, y los no semejantes son {no_comunes}')
```

```python
#Ejercicio 12
textos = ['hola', '', 'python', '', 'listas', ' ', 'programar', '']
limpios= [
    texto
    for texto in textos
    if texto.strip()
]
print(f'limpios: {limpios}')

despejado=[]
for palabra in textos:
        palabra.strip
        despejado.append(palabra)
print(f'Lista limpia: {despejado}')
```

```python
#Ejercicio 16
numeros = [10, 20, 30, 40, 50, 60, 70, 80]
primera_mitad=[]
segunda_mitad=[]
long_lista=len(numeros)
indice_mitad=long_lista/2
for idx, numero in enumerate(numeros):
    if idx<indice_mitad:
        primera_mitad.append(numero)
    else:
        segunda_mitad.append(numero)
print(f'Primera mitad: {primera_mitad}')
print(f'Segunda mitad: {segunda_mitad}')

```
