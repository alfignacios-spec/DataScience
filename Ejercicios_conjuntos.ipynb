```python
#ejercicio 1
estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"} 
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}

estudiantes_ambos=estudiantes_javascript&estudiantes_python
print('Estudiantes en ambos cursos: ',estudiantes_ambos)

solo_python=estudiantes_python.difference(estudiantes_javascript)
solo_java=estudiantes_javascript.difference(estudiantes_python)
todos_estudiantes=estudiantes_python.union(estudiantes_javascript)

print('Estudiantes solo de python: ', solo_python)
print('Estudiantes solo de javascript: ', solo_java)
print('Todos lo estudiantes son: ', todos_estudiantes)
```

```python
#ejercicio 2
almacen_arequipa = { "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos" } 
almacen_lima = { "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes" }

productos_ambos=almacen_arequipa&almacen_lima
print('Los articulos que tienen ambos almacenes: ', productos_ambos)

solo_aqp=almacen_arequipa-almacen_lima
print('Los productos que solo estam disponibles en Arequipa: ', solo_aqp)

solo_lima=almacen_lima-almacen_arequipa
print('Los productos que solo estam disponibles en Lima: ', solo_lima)


todos_productos=almacen_arequipa|almacen_lima
print('Todos los productos disponibles son: ', todos_productos)

que_fue=almacen_arequipa==almacen_lima
if que_fue:
    print('Los dos almacenes cuentan con el mismo inventario')
else:
    print('Los almacenes no cuentan con el mismo inventario')

```

```python
#ejercicio 3
permisos_usuario = { "leer", "editar", "comentar", "descargar" } 
permisos_requeridos = { "leer", "editar", "eliminar", "publicar" }

usuario_apto=permisos_usuario.issubset(permisos_requeridos)
print(usuario_apto)
usuario_apto: bool
if usuario_apto:
    print('El usuario cuenta con todos los permisos necesarios')
else:
    print('El usuario no cuenta con todos los permisos')

permisos_faltantes=permisos_requeridos-permisos_usuario
print(f'Al usuario le faltan los siguientes permisos: {permisos_faltantes}')

permisos_excecivos=permisos_usuario-permisos_requeridos
print(f'Al usuario cuenta adicionalmente con los permisos: {permisos_excecivos}')


permisos_diferentes=permisos_usuario^permisos_requeridos
lista_usuario=list(permisos_diferentes)
n_permisos=len(lista_usuario)

print(f'El usuario cuenta con {n_permisos} permiso(s) diferente(s).')
```

```python
#ejercicio 4
numeros_a = {2, 4, 6, 8, 10, 12, 14} 
numeros_b = {4, 6, 9, 10, 13, 14, 16} 
numeros_c = {1, 4, 6, 10, 11, 14, 18}

numeros_comunes=numeros_c&numeros_a&numeros_b
print('Los numeros en comun de los 3 conjuntos es: ', numeros_comunes)

elementos1=numeros_a&numeros_b-numeros_c
print('Los elementos presentes en numeros_a y numeros_b, pero no en numeros_c: ',elementos1)

print('Los numeros que solo son de a: ', numeros_a-(numeros_b|numeros_c))

print('Los numeros unicos son: ',(numeros_c^numeros_a)|(numeros_a^numeros_b)|(numeros_c^numeros_b))
```

```python
#ejercicio 5
habilidades_candidato_1 = {"Python", "Git", "SQL", "Docker", "React"} 
habilidades_candidato_2 = {"Python", "JavaScript", "React", "Node.js", "Git"} 
habilidades_requeridas = {"Python", "Git", "SQL", "Docker"} 

print('Las habilidades en comun de los candidatos: ', habilidades_candidato_1&habilidades_candidato_2)

print('Las habilidades exclusivas del candidato 1: ',habilidades_candidato_1-habilidades_candidato_2)
print('Las habilidades exclusivas del candidato 2: ',habilidades_candidato_2-habilidades_candidato_1)

print('Las habilidades que le faltan al candidato 1: ',habilidades_requeridas-habilidades_candidato_1)
print('Las habilidades que le faltan al candidato 2: ',habilidades_requeridas-habilidades_candidato_2)

print('Todas las habilidades serian: ', habilidades_requeridas|habilidades_candidato_1|habilidades_candidato_2)
```
