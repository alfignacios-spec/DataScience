#condicionales
-si se cumple esta condicion hago:
proceso 1
- si se cumple esta otra:
proceso 2
- si no se cumple ninguna  de las anteriores:
proceso por default

```python
#condicionales
edad=int(input('Ingresa tu edad: '))
if edad>=40:
    print('Estoy viejo :,v')
elif edad>=18:
    print('Soy un adulto')

elif edad>=13:
    print('Soy un adolescente')

else:
    print('Soy un infante')
```
