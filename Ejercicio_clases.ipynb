```python
#Ejercicio 1
class Persona:
    def __init__(self,nombre:str,edad:int)->None: #Siempre va self, creamos los parametros (elementos) nombre y edad
        self.nombre=nombre
        self.edad=edad

    def saludar(self):
        print(f'Hola soy {self.nombre} y tengo {self.edad} años')

jorge=Persona('Jorge',10) #creamos objetos variados
maria=Persona('Maria',20)
luis=Persona('Luis',15)

jorge.saludar()
maria.saludar()
luis.saludar()
```

```python
#Ejercicio 2
class Rectangulo:
    def __init__(self,base:float, altura:float)->None: #definimos una metodo iniciador
        self.base=base #definimos atributos
        self.altura=altura

    def area(self)->float: #definimos otros metodos, deben tener siempre la misma estructura de self
        return self.base*self.altura
    
    def perimetro(self)->float:
        return 2*(self.base+self.altura) #return porque aun trabajremos con este dato, no queremos solamente visualizarlo
    
    def __str__(self)->str: #el metodo str te devuelve un print
        print(f'Rectangulo base: {self.base} y altura: {self.altura}') #print, porque solo queremos un mensaje de salida, no utilizaremos mas este dato
        
mi_rectangulo=Rectangulo(3,5)
print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())
print(mi_rectangulo)
```
