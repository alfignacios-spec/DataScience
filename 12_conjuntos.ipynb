```python
#sets (conversion a conjunto)
conjunto_a={1, 3, 5, 7, 9, 1, 3, 9}
conjunto_b={3, 5, 7, 562}
conjunto_c=[1, 3, 5, 7, 9, 1, 3, 9]

conjunto_c[0]=30
print(conjunto_c)

print(conjunto_a.union(conjunto_b))
print(conjunto_a.intersection(conjunto_b))
print(conjunto_a.difference(conjunto_b))
print(conjunto_a.symmetric_difference(conjunto_b))

print(conjunto_a|conjunto_b)
print(conjunto_a&conjunto_b)
```

```python
lista=[3,3,3,4,5,6,6]
conjunto_nuevo=set(lista)
print(conjunto_nuevo)
```

```python
conjunto={1,2,3}
conjunto.add(9)
print(conjunto)

conjunto.remove(3)
print(conjunto)

conjunto.pop()
print(conjunto)

lista=[3,3,3,4,5,6,6]
lista.insert(0,0)
print(lista)

#para ver si no hay elementos en comun
print(conjunto_a.isdisjoint(conjunto_b))


#para ver si un conjunto es subconjunto de otro
conjunto_a={1, 3, 5, 7, 9, 1, 3, 9}
conjunto_b={3, 5, 7}

print(conjunto_b.issubset(conjunto_a))

#para ver si es superconjunto
print(conjunto_a.issuperset(conjunto_b))
```
