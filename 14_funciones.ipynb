```python
def es_par(numero:int)->bool:
    if numero%2==0:
        return True
    return False

print(es_par(19))
print(es_par(10))

def mayor(numero1:int | float, numero2:int|float)->int|float:
    if numero1>numero2:
        return numero1
    return numero2
print(mayor(10,10))
print(mayor(11,56))

def promedio(valores:list[int|float])->float:
    return sum(valores)/len(valores)
lista=[2,3,4,5]
print(promedio(lista))

def contar_vocales(texto:str)->int:
    vocales=['a','e','i','o','u']
    contador=0
    for letra in texto:
        if letra in vocales:
            contador+=1
    return contador

print(contar_vocales('La casa esta vacia'))
```
