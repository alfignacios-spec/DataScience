```python
#Ejercicio 1
inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3} 
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5} 
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

inv_total=dict(inventario_norte)
#tambien se pudo inicializar como inv_total=dict(inventario_norte), y asi solo hacias 2 ciclos for
#dict: permite que no mute o se vea afectado el diccionario original ante un cambio posterior (crea una copia nueva)

for llave, valor in inventario_sur.items(): #con el .items se accede al contenido de los dicc
    print(f'Llave:{llave}-valor:{valor}')
    if llave not in inv_total:
        inv_total[llave]=0 #se esta creando una llave en caso no exista, inicia en 0
    inv_total[llave]+=valor #en caso ya exista, se suma el 'valor' (numeros en cada tupla)
    print(f'inventario total-llave:{inv_total[llave]}') #se especifica que imprima la llave
    print('='*10) #hace enter 

```

```python
for llave, valor in inventario_centro.items(): #con el .items se accede al contenido de los dicc
    print(f'Llave:{llave}-valor:{valor}')
    if llave not in inv_total:
        inv_total[llave]=0 #se esta creando una llave en caso no exista, inicia en 0
    inv_total[llave]+=valor #en caso ya exista, se suma el 'valor' (numeros en cada tupla)
    print(f'inventario total-llave:{inv_total[llave]}') #se especifica que imprima la llave
    print('='*10) #hace enter 


print(inv_total)

ordenadas=sorted(inv_total.keys())
print(ordenadas)

#nuevo diccionario
inventario_ordenado=[]
for llave in ordenadas:
    inventario_ordenado[llave]=inv_total[llave]
print(inv_total)
print(inventario_ordenado)
```

```python
calificaciones = { "Ana": [17, 18, 16, 19], 
                  "Luis": [15, 16, 17, 16], 
                  "Carla": [18, 17, 19, 18], 
                  "Pedro": [16, 16, 16, 16], 
                  "Lucía": [15, 17, 16, 16] }

estadisticas={}
for estudiante,notas in calificaciones.items(): #.items permite recorrer cada diccionario
    print(estudiante,notas)
    estadisticas[estudiante]={ #se crea la llave estudiantes, por ejemplo Ana, Luis, etc
        'minima nota':min(notas), #asignamos diccionarios de acuerdo a cada llave:
        'maxima nota':max(notas),
        'promedio de notas':sum(notas)/len(notas)
    }
    print(estadisticas[estudiante])#imprime las estadisticas obtenidas para cada estudiante
    print('='*20) #imprime la separacion entre cada diccionario
```

```python
for estudiante, stats in estadisticas.items():
    print(estudiante)
    print(stats)
    print('='*20)
```

```python
ranking=[]
for nombre, stats in estadisticas.items(): #debemos generar tuplas
    ranking.append((nombre,stats['promedio de notas'])) #se ingresan tuplas con append

ranking=sorted(ranking, key=lambda x:x[1], reverse=True) #key=como lo ordeno, extraes de la tupla, el valor numero 1 para ordenarlo
#cada 'x' significa elemento: y se va a tomar la posicion numero 1 de cada elemento (considera que la posicion cero en si es la primera)
#reverse es verdadero=ordena ascendentemente
#si reverse es false, osea usamos reverse: ordena descendentemente
#la funcion key permite este tipo de configuracion


#tenemos 3ple empate, nos situamos en el tercer elemento, y a partir de ahi volvermos a ordenar:
#ya no ordenaremos por el elmento '1', sino por el '0', osea los nombres
ranking[2:]=sorted(ranking[2:], key=lambda x:x[0], reverse=False)#nos situamos en el tercer elemento, ya que aqui tenemos conflicto
#se utiliza el x=0porque ya no quiero ordenar por notas (elemento 1), sino por el nombre (el primer elemento)
#finalmente, utilziamos reverse false debido a que lo queremos alfabeticamente en este caso
print(ranking)
```

```python
#Ejercicio 4
texto = "Python es poderoso; Python es flexible, y aprender Python puede ser muy divertido." 
palabras_vacias = {"es", "y", "ser", "muy", "puede"}
contador={}
for palabra in texto.split(): #.split para especificar que considere palabras, y no caracteres
    limpio=palabra.lower().replace(';','').replace(';','').replace('.','') #lower vuelve todo en minuscula
    
    if limpio in palabras_vacias:#para limpiar los espacios
        continue
   
    if limpio not in contador:
        contador[limpio]=0
    contador[limpio]+=1
print(contador)
```

```python
#ejercicio 5
empleados = [ 
{"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario": 
5200}, 
{"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800}, 
{"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario": 
3400}, 
{"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario": 
5600}, 
{"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700} 
] 

resumen={}
for idx, empleado in enumerate(empleados):
    
    departamento=empleado['departamento']
    nivel=empleado['nivel']
    nombre=empleado['nombre']
    salario=empleado['salario']
    
    if empleado['departamento'] not in resumen:
        resumen[departamento]={}

    if nivel not in resumen[departamento]:

        resumen[departamento][nivel]={
            'cantidad':0,
            'nombres':[],
            'salarios':[]
        }

    grupo=resumen[departamento][nivel]
    grupo['nombres'].append(nombre)
    grupo['salarios'].append(salario)
    grupo['cantidad']+=1
    resumen[departamento][nivel]['cantidad']+=1
    resumen[departamento][nivel]['nombres'].append(nombre)
    resumen[departamento][nivel]['salarios'].append(salario)
    


for registro,grupo_xp in resumen.items():
    for experiencia, valor in grupo_xp.items():
        valor['promedio']=sum(valor['salarios'])/len(valor['salarios'])
        
print(resumen)
```

```python
#ejercicio 6
tienda_a = {"laptop": 3500, "mouse": 80, "teclado": 160, "monitor": 920} 
tienda_b = {"laptop": 3400, "mouse": 90, "teclado": 150, "webcam": 220} 

productos=list(tienda_a.keys())
productos.extend(list(tienda_b.keys()))
productos_unicos=set(productos)
print(productos_unicos)

comparacion={}
for var in productos_unicos:
    if var in tienda_a and var in tienda_b:
        comparacion[var]={
        'diff':abs(tienda_a[var]-tienda_b[var]),
        'mas barato':'tienda a' if tienda_a[var]<tienda_b[var] else 'tienda b',
        'precio': tienda_a[var] if tienda_a[var]<tienda_b[var] else tienda_b[var]
        }

    if var in tienda_a and var not in tienda_b:
        comparacion[var]={
        'diff':'No hay diferencia',
        'mas barato':'tienda a',
        'precio': tienda_a[var]
        }
    
    if var in tienda_b and var not in tienda_a:
        comparacion[var]={
        'diff':'No hay diferencia',
        'mas barato':'tienda b',
        'precio': tienda_b[var]
        }
for comp in comparacion:
    print(comp)
    print(comparacion[comp])
    print('='*100)
print(comparacion)    
```

```python
#Ejercicio 7
candidatos = {"Ana", "Luis", "Carla"} 
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapeo={
    'ana':'Ana',
    'luis':'Luis',
    'carla':'Carla'
}
conteo={}
for voto in votos:
    limpio=voto.lower()
    candidato=mapeo.get(limpio,'invalido')

    if candidato=='invalido':
        conteo.setdefault('Votos invalidos',0)
        conteo['Votos invalidos']+=1
        continue
    
    conteo.setdefault(candidato,0)
    conteo[candidato]+=1

total=len(votos)
for candidato in candidatos:
 conteo[f'{candidato}_pct']=round(conteo[candidato]/total *100,2)

conteo['Votos invalidos_pct']=round(conteo['Votos invalidos']/total*100,2)
print(conteo)

```
